```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from xgboost import XGBRegressor
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
```

```python
data = pd.read_csv('train.csv')
```

```python
data['date'] = pd.to_datetime(data['date'], format='%d-%m-%Y')
data.set_index('date', inplace=True)
```

```python
item_id = 1
store_id = 1
data_item_store = data[(data['item'] == item_id) & (data['store'] == store_id)]
data_item_store = data_item_store['sales']
```

```python
train, test = train_test_split(data_item_store, test_size=0.2, shuffle=False)
```

```python
# Create features for XGBoost
def create_features(df):
    df['year'] = df.index.year
    df['month'] = df.index.month
    df['day'] = df.index.day
    return df
```

```python
train_xgb = create_features(train.to_frame())
test_xgb = create_features(test.to_frame())
```

```python
# XGBoost model
xgb = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=5)
xgb.fit(train_xgb[['year', 'month', 'day']], train)
xgb_predictions = xgb.predict(test_xgb[['year', 'month', 'day']])
xgb_rmse = np.sqrt(mean_squared_error(test, xgb_predictions))
```

```python
# Prepare data for LSTM
def create_lagged_features(data, n_lag=3):
    df = pd.DataFrame(data)
    columns = [df.shift(i) for i in range(1, n_lag + 1)]
    df = pd.concat(columns, axis=1)
    df.dropna(inplace=True)
    return df
```

```python
n_lag = 3
train_lstm = create_lagged_features(train, n_lag)
test_lstm = create_lagged_features(test, n_lag)
```

```python
train_X, train_y = train_lstm.iloc[:, :-1].values, train_lstm.iloc[:, -1].values
test_X, test_y = test_lstm.iloc[:, :-1].values, test_lstm.iloc[:, -1].values
```

```python
train_X = train_X.reshape((train_X.shape[0], train_X.shape[1], 1))
test_X = test_X.reshape((test_X.shape[0], test_X.shape[1], 1))
```

```python
lstm = Sequential()
lstm.add(LSTM(50, input_shape=(train_X.shape[1], train_X.shape[2])))
lstm.add(Dense(1))
lstm.compile(loss='mean_squared_error', optimizer='adam')
lstm.fit(train_X, train_y, epochs=20, batch_size=1, verbose=2)
lstm_predictions = lstm.predict(test_X)
lstm_rmse = np.sqrt(mean_squared_error(test_y, lstm_predictions))
```

```python
print(f'XGBoost RMSE: {xgb_rmse}')
print(f'LSTM RMSE: {lstm_rmse}')
```

```python
aligned_index = test.index[-len(lstm_predictions):]
```

```python
plt.figure(figsize=(12, 6))
plt.plot(test.index, test, label='Actual Sales')
plt.plot(test.index, xgb_predictions, label='XGBoost Predictions')
plt.plot(aligned_index, lstm_predictions, label='LSTM Predictions')
plt.legend()
plt.show()
```
